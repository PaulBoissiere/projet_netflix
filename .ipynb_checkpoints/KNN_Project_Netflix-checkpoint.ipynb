```python
weights['averageRating'] = 3
weights.head(30)
```

```python
movie = input('Insere un film')
def recommandation2(movie):
    if df_scaled[df_scaled['title'].str.contains(movie)].shape[0] > 1:
        print("Plusieurs films contiennent ce nom, choisissez parmi l'un d'eux : ")
        return df_final_knn[df_final_knn['title'].str.contains(movie)]
    else:
        model_film = NearestNeighbors(metric = 'wminkowski', n_neighbors=10, metric_params = {"w": weights}).fit(X_scaled)
        index_reco = model_film.kneighbors(df_scaled.loc[df_scaled['title'].str.contains(movie), X.columns])
        recommended_movie = df_final_knn.iloc[df_final_knn.index.isin(index_reco[1][0][1:9])]
        return recommended_movie
recommandation2(movie)
```
