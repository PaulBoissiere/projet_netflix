```python
from sklearn.neighbors import KNeighborsClassifier, NearestNeighbors
```

```python
X = df_knn[['averageRating', 'numVotes', 'action', 'adventure', 'animation', 'biography', 'comedy',
       'crime', 'documentary', 'drama', 'family', 'fantasy', 'film-noir',
       'history', 'horror', 'music', 'musical', 'mystery', 'news', 'romance',
       'sci-fi', 'sport', 'thriller', 'war', 'western', '1930', '1940', '1950',
       '1960', '1970', '1980', '1990', '2000', '2010', '2020']]
```

```python
model_film = NearestNeighbors(metric = 'minkowski', n_neighbors=10, ).fit(X)
```

```python
model_film.kneighbors(df_knn.loc[df_knn['title'].str.contains("Mission C"), ['averageRating', 'numVotes', 'action', 'adventure', 'animation', 'biography', 'comedy',
       'crime', 'documentary', 'drama', 'family', 'fantasy', 'film-noir',
       'history', 'horror', 'music', 'musical', 'mystery', 'news', 'romance',
       'sci-fi', 'sport', 'thriller', 'war', 'western', '1930', '1940', '1950',
       '1960', '1970', '1980', '1990', '2000', '2010', '2020']])
```

```python
movie = input('Insere un film')
def recommandation2(movie):
    if df_knn[df_knn['title'].str.contains(movie)].shape[0] > 1:
        print("Plusieurs films contiennent ce nom, choisissez parmis l'un d'eux : ")
        return df_knn[df_knn['title'].str.contains(movie)]
    else:
        index_reco = model_film.kneighbors(df_knn.loc[df_knn['title'].str.contains(movie), ['averageRating', 'numVotes', 'action', 'adventure', 'animation', 'biography', 'comedy','crime', 'documentary', 'drama', 'family', 'fantasy', 'film-noir','history', 'horror', 'music', 'musical', 'mystery', 'news', 'romance','sci-fi', 'sport', 'thriller', 'war', 'western', '1930', '1940', '1950','1960', '1970', '1980', '1990', '2000', '2010', '2020']])
        recommended_movie = df_knn.iloc[df_knn.index.isin(index_reco[1][0][1:9])]
        return recommended_movie
```
